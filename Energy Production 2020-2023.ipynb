```python
import pandas as pd
import numpy as np
```

```python
EnergyProducers=pd.read_excel('MyEnergyData.xlsx', sheet_name='Million tonnes')
```

```python
EnergyProducers.head()
```

```python
 EnergyProducers.info()
```

# Top 10 countries by Energy Production

```python

total_by_country = EnergyProducers.groupby('Country')['Total Production'].sum().reset_index()
top_10 = total_by_country.sort_values(by='Total Production', ascending=False).head(10)
top_10.reset_index(drop=True, inplace=True)
top_10.index = top_10.index + 1
print(top_10)


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(top_10['Country'], top_10['Total Production'], color='skyblue')
plt.title('Top 10 Countries by Total Energy Production')
plt.xlabel('Country')
plt.ylabel('Total Energy Production')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Leader in energy self-sufficiency

```python
df=EnergyProducers
```

```python
import numpy as np

df['Self_Sufficiency'] = np.where(
    df['Total Consumption'] < 1,  # treat anything <1 as zero
    0,  # or some other default
    df['Total Production'] / df['Total Consumption']
)

```

```python
leader = df.loc[df['Self_Sufficiency'].idxmax(), ['Country', 'Self_Sufficiency']]
print(leader)
```

# Top 10 countries by energy self-sufficiency


```python
self_sufficiency_by_country = df.groupby('Country')['Self_Sufficiency'].sum().reset_index()
top_10 = self_sufficiency_by_country.sort_values(by='Self_Sufficiency', ascending=False).head(10)
top_10.reset_index(drop=True, inplace=True)
top_10.index = top_10.index + 1
top_10['Self_Sufficiency'] = top_10['Self_Sufficiency'].round(2)
print(top_10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_10['Country'], top_10['Self_Sufficiency'], color='orange')
plt.title('Top 10 Countries by Energy Self-Sufficiency')
plt.xlabel('Country')
plt.ylabel('Self-Sufficiency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
